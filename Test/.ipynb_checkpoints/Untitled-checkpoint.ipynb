```python
import numpy as np
import pandas as pd
```

```python
def randChar(f, numGrp, N):
    things = [f%x for x in range(numGrp)]
    return [things[x] for x in np.random.choice(numGrp, N)]

def randFloat(numGrp, N):
    things = [round(100*np.random.random(),4) for x in range(numGrp)]
    return [things[x] for x in np.random.choice(numGrp, N)]
```

```python
N = int(1e6)
K = 100
```

```python
%%time
df = pd.DataFrame({
    'id1': randChar('id%03d', K, N),
    'id2': randChar('id%03d', K, N),
    'id3': randChar('id%010d', N//K, N),
    'id4': np.random.choice(K, N),
    'id5': np.random.choice(K, N),
    'id6': np.random.choice(N//K, N),
    'v1': np.random.choice(5, N),
    'v2': np.random.choice(5, N),
    'v3': randFloat(100, N)
})
```

```python
df.head()
```

```python
%time df.groupby(['id1']).agg({'v1': sum})
```

```python
%time df[['id1', 'v3']].groupby(['id1']).rank(method='max', numeric_only=True, pct=True)
```
